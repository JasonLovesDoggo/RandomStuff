1. Internet Usage and Communication Comfort


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data for Internet Usage and Communication Comfort
data = {
    'Internet Usage': ['1-2 hours', '3-4 hours', '5-6 hours', 'More than 6 hours'],
    'Comfort Level': [4, 3, 2, 1]  # Scale: 1 (Strongly Disagree) to 5 (Strongly Agree)
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.barplot(x='Internet Usage', y='Comfort Level', data=df, palette='viridis')
plt.title('Internet Usage vs. Communication Comfort')
plt.xlabel('Internet Usage')
plt.ylabel('Comfort Level')
plt.ylim(0, 5)
plt.xticks(rotation=45)
plt.show()
```

3. Social Media and Emotional Expression


```python
# Sample data for Social Media and Emotional Expression
labels = ['Yes', 'No', 'Maybe']
sizes = [60, 20, 20]  # Example percentages

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#66c2a5','#fc8d62','#8da0cb'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Social Media Makes It Easier to Express Emotions')
plt.show()
```

5. Engagement in Filler Conversations


```python
# Sample data for Engagement in Filler Conversations
filler_data = {
    'Frequency': ['Rarely', 'Sometimes', 'Often'],
    'Agree': [20, 50, 80],  # Percentage of students who agree
    'Disagree': [80, 50, 20]  # Percentage of students who disagree
}

filler_df = pd.DataFrame(filler_data)

filler_df.set_index('Frequency').plot(kind='bar', stacked=True, figsize=(8, 5), color=['#66c2a5', '#fc8d62'])
plt.title('Engagement in Filler Conversations vs. Perception of Social Media')
plt.xlabel('Frequency of Filler Conversations')
plt.ylabel('Percentage of Students')
plt.legend(title='Perception', labels=['Agree', 'Disagree'])
plt.xticks(rotation=0)
plt.show()
```

6. Perceived Benefits vs. Drawbacks


```python
# Sample data for Perceived Benefits vs. Drawbacks
benefits_data = {
    'Response': ['Benefits Outweigh Drawbacks', 'Drawbacks Outweigh Benefits', 'Neutral'],
    'Percentage': [50, 30, 20]  # Example percentages
}

benefits_df = pd.DataFrame(benefits_data)

plt.figure(figsize=(8, 5))
sns.barplot(x='Percentage', y='Response', data=benefits_df, palette='pastel')
plt.title('Perceived Benefits vs. Drawbacks of Social Media')
plt.xlabel('Percentage of Students')
plt.ylabel('Response')
plt.xlim(0, 100)
plt.show()
```
